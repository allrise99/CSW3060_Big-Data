## 연습문제 1

```python
import numpy as np
import pandas as pd

a = pd.read_csv('Regularity_DF.txt', sep='\t', index.col = 0)
```

## 차트박스 밖 출력

```python
import matplotlib.pyplot as plt

a.plot()
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5))

#center : x 좌표 기준 왼쪽 정렬되도록 함
#bbox to anchor (0.5, 1) -> 좌표값 기준 center left
```

## 연습문제 2

```python
# , : 전체 column
b = a.loc[1800:1819,].median(axis = 0)
c = a.loc[2000:2019,].median(axis = 0)
```

## 연습문제 3

```python
d = pd.DataFrame({'1800~1819':b, '2000~2019':c})
```

## scatter plot

```python
plt.scatter(d['1800~1819'], d['2000~2019'])
plt.plot([0, 1], [0, 1], linestyle = 'dashed')
plt.xlabel('Median Regularity 1800~1819')
plt.ylabel('Median Regularity 2000~2019')

# x-y 선에 가까우면 변화가 없다고 볼 수 있음
```

```python
for i in d.index :
    print(i, np.array(d.loc[i,]))
```

```python
for i in d.index :
    plt.annotate(i, np.array(d.loc[i,]), size = 11, rotation=20)
plt.plot([0, 1], [0, 1], linestyle='dashed')
plt.xlabel('Median Regularity 1800~1819')
plt.ylabel('Median Regularity 2000~2019')
```
