```python
x = np.array([1.1, 1.2, 1.3, 1.4])
y = np.array([2.1, 2.2, 2.3, 2.4])
condition = np.array([True, True, False, False])

np.where(condition, x, y)
#condition : True / False 로 출력하는 데이터
#True의 경우 x (첫번째 부분) / False의 경우 y (두번째 부분) 출력
```
