```python
from GoogleNgrams import Ngrams
Ngrams('burned', start = 1800, end = 2019, language = 'eng_2019')
```

```python
a = Ngrams('Charles Dickens, Merry Christmas')
a.index.name = 'YEAR'
a.columns.name = 'KEYWORD'
a
```

```python
import matplotlib.pyplot as plt
a.plot()
```

```python
a.plot()
plt.title('Writer vs Greeting')
plt.ylabel('Relative Frequency')
plt.xlabel(a.index.name.lower())
plt.legend(title = a.columns.name.lower())
```

```python
a.plot()
plt.axis([1820, 1950, 0, a['Charles Dickens'].max()])

#a.columnname은 컬럼 이름에 공백이 포함된 경우 사용 X
```

```python
import numpy as np
a.loc[1820:].plot()
plt.xticks(range(1820, 2019, 30), rotation = 45)
plt.yticks(np.arange(0, 1.5e-6, 0.5e-6), rotation = 25)
```

```python
a.plot()
plt.xticks([1812, 1836, 1843, 1870], 
           ['출생', '출세작 출간', 'Christmas Carol 출간', '사망'], fontsize = 'medium')
```
