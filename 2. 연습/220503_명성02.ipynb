## Series 문자열 연결

```python
import pandas as pd

b = pd.Series(a).str.cat(['fre_2019', 'rus_2019'], sep=':')
b
```

```python
from GoogleNgrams import Ngrams
import matplotlib.pyplot as plt

terms = pd.Series(a).str.cat(['fre_2019', 'rus_2019'], sep=':')
b = Ngrams(', '.join(terms), start = 1900, end = 2000)
# ', '.join(terms)를 통해 Marc Chagall:fre_2019, Mapk ..:rus_2019 형태로 나옴
b.plot()
```

```python
b.columns
```

```python
#regex : regular expression
# .(마침표) / + 선행하는 문자가 1회 이상 반복
# .+ : 하나 이상 연쇄되는 것을 비문자열로 바꿔라 
#
b.columns = b.columns.str.replace(':.+', '', regex = True) 
b.head()
```

```python
b.plot()
plt.xticks([1900, 1910, 1917, 1923, 1941, 1950],
           [1900, '프랑스유학', '러시아예술감독', '프랑스귀화', '미국이주', 1950], 
           rotation = 80, fontsize = 'large')
plt.title('Marc Chagall')
plt.xlabel('Year')
plt.ylabel('Relative Frequency')
```
