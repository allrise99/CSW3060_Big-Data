```python
import pandas as pd

a = pd.read_csv('IrregularVerbs.txt', sep = '\t', index_col = 0)
a
```

```python
from GoogleNgrams import Ngrams
b = Ngrams(', '.join(a['Regular']))
b.shape
```

```python
b.iloc[:5, :6]
```

```python
new_index = pd.MultiIndex.from_product([['Regular'], a.Regular[:]], names = ['Type', 'Past'])
new_index
```

```python
b.columns = pd.MultiIndex.from_product([['Regular'], a.Regular[:]], names = ['Type', 'Past'])
b
```

```python
ex_prod = [['a', 'b', 'c', 'd'], ['a', 'b']]
ex_index = pd.MultiIndex.from_product(ex_prod, names = [1, 2])
print(ex_index)
```

```python
c = Ngrams(', '.join(a['Irregular']))
c
```

```python
c.columns = pd.MultiIndex.from_product([['Irregular'], a.Irregular[:]], names = ['Type', 'Past'])
c.iloc[:5, :6]
```

```python
d = pd.concat([b, c], axis = 1)
d.iloc[:5, [0, 1, 12, 13]]
```

```python
d.to_csv('Irregular_DF.txt', sep = '\t')
```
