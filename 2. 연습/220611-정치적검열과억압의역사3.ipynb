```python
plt.plot(a, color = 'gray', linewidth = 1, alpha = 0.8, label = '_nolegend_')
plt.plot(a.median(axis = 1), linewidth = 3, color = 'black', label = 'median')
# a.median(axis = 1)을 통해 median 값 표시

plt.vlines(x = 1950, ymin = 0, ymax = np.array(a).max(), 
           linewidth = 0.8, color = 'red', linestyle = 'dashed')
plt.xticks([1910, 1947, 1950, 1960, 1970], rotation = 45, horizontalalignment = 'right')
plt.legend()
```

## 연습문제 6

```python
import math, time
file = pd.read_csv('Fame.txt', sep = '\t')
dfs = []
for i in range(0, (math.ceil(len(file.Name)/12))) :
    dfs.append(Ngrams(', '.join(file.Name[(i*12) : ((i+1)*12)]), 
                      start = 1946, end = 1965,
                      language = 'eng_us_2019'))
    time.sleep(0.5)
a = pd.concat(dfs, axis = 1)
```

```python
a.columns = pd.MultiIndex.from_product((['Fame'], a.columns), names = ['Category', 'Name'])
a.head()
```

## 연습문제 7

```python
file = open('US_Suppression.txt').read().strip().split('\n')
df = Ngrams(', '.join(file), start = 1946, end = 1965, 
            language = 'eng_us_2019')
df.columns = pd.MultiIndex.from_product((['US Blacklists'], df.columns),
                                        names = ['Category', 'Name'])
b = pd.concat([a, df], axis = 1)
```

## 연습문제 8

```python
b.head()
```

```python
b['Fame'].median(axis = 1).pct_change().hist(alpha = 0.5)
b['US Blacklists'].median(axis = 1).pct_change().hist(alpha = 0.5)
plt.legend(b.columns.levels[0])
```

```python
c = np.log(b.loc[(1950<b.index) & (b.index < 1961), ].median(axis = 0) / 
           b.loc[~((1950 < b.index) & (b.index < 1961)), ].median(axis = 0))
c.head()
```

```python
c['Fame'].hist(alpha = 0.5)
c['US Blacklists'].hist(alpha = 0.5)
plt.legend(c.index.levels[0])
```
