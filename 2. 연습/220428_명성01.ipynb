```python
from GoogleNgrams import Ngrams
```

```python
a = Ngrams('Charles Dickens, Merry Christmas')
a.index.name = 'YEAR'
a.columns.name = 'KEYWORD'
```

```python
import matplotlib.pyplot as plt
a.plot()
```

```python
a.plot()
plt.xticks([1812, 1836, 1843, 1870], 
           ['출생', '출세작 출간', 'Christmas Carol 출간', '사망'], fontsize = 'medium')
plt.annotate('출생', xy = (1812, a.loc[1812, 'Charles Dickens']), 
             xytext = (1812, 6.0e-7), horizontalalignment = "right", 
             arrowprops = dict(arrowstyle = '->'))

plt.xlim(1810, 1900)
```
