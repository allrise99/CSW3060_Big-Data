## 연습문제 1

```python
import pandas as pd
from GoogleNgrams import Ngrams

file = open('Chagall.txt', encoding="cp949").read().split(',')
terms = pd.Series(file[0:1]*2 + file[1:]).str.cat(['fre_2019', 'eng_us_2019', 'rus_2019'], sep=':')

a = Ngrams(', '.join(terms), start = 1900, end = 1960)
a
#a.columns = a.columns.str.replace('.+:|_\d+', '', regex = True).str.upper()
```

```python
#리스트를 slicing 한 결과는 리스트 
file[0:1]*2 + file[1:]
```

```python
#리스트를 indexing하면 해당 원소만 나옴
file[0] * 2 + file[1]
```

```python
#Series에 대해서 적용을 해줄때는 
#해당 Series 변수 뒤에 .str을 붙인 다음 쓸 메소드를 적용한다.

# (.+:) |(또는) (_\d+) : 'Marc Chagall:' 또는 '_2019' 둘 다를 없앰
a.columns = a.columns.str.replace('.+:|_\d+', '', regex = True).str.upper()
a
```

## 연습문제 6 

```python
for i, j in [('fre_2019', 'French'), ('eng_us_2019', 'American English')] :
    print('code : ', i, 'corpus : ', j)
```

```python
file2 = open('EU_painters.txt').read().split('\n')
dfs = [] #두 데이터를 합칠 빈 DataFrame
for i, j in [('fre_2019', 'French'), ('eng_us_2019', 'American English')] :
    df = Ngrams(', '.join(file2), start = 1900, end = 1960, language = i)
    df.columns = pd.MultiIndex.from_product([[j], df.columns], 
                                            names = ['Corpus', 'Painter'])
    dfs.append(df)
b = pd.concat(dfs, axis = 1) #메소드가 아니므로 치환문을 사용해야 함
b
```

## 연습문제 7

```python
c = pd.concat([a, b], axis = 1)
c
```

- Marc Chagall이 세 번 반복되는데 해당하는 index 값을 세개 입력해도 되나 c.loc[:, (slice(None), 'Marc Chagall')].plot() 형태도 가능

- [:, ] : 행 인덱스 전체 가리킴

- (slice(None), 'Marc Chagall') : (최상위 계층, 하위 계층) 으로 구성
    - 튜플 형태로 입력
    - **slice(None) : 최상위 계층에 관계없이**
    - 세 번 출력하되 최상위 계층에 관계 없이
