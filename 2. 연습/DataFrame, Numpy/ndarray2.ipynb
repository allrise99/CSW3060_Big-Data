```python
a = np.arange(1, 7) # [1, 2, 3, 4, 5, 6]
print(a[5]) # 6
print(a[2:4]) # [3, 4]
a[5] = 12 
print(a) #[1, 2, 3, 4, 5, 12]
```

```python
a = np.arange(1, 10, 2)
print(a) # [1, 3, 5, 7, 9]
print(a[::]) #동일
print(a[::2]) # [1, 5, 9]
print(a[::-2]) # [9, 5, 1]
print()
print(a[-2::-1]) # 앞의 -2 : 뒤에서 두 번째 의미
print(a[2::-1]) # [5, 3, 1]
print(a[-3::-1])
```

```python
a = np.arange(1, 13).reshape(3, 4)
print(a)
print()
print(a[-1])    # a 행렬에서 가장 마지막 행을 출력해라 -> [9, 10, 11, 12]
print(a[0][-1]) # 0번째 행 마지막 원소 출력 -> [[4]]가 아니다!! 4
print(a[-1][0]) # 마지막 행 첫번째 원소 출력 -> 9
print()
print(a[0:2])   # 0 ~ 1행 출력 : 2차원 형태로 출력!!!
print(a[0:1])   # 0행 출력 : 2차원 형태로 동일하게 출력
print(a[0:2][0]) # 0행 출력 : 1차원 형태로 출력
```

## 2차원 ndarray
ex. a = [[1, 2, 3, 4], 
         [5, 6, 7, 8]]

- a[0] -> [1, 2, 3, 4] : 0번째 행 출력, 1차원 형태로 출력
- a[0][0] -> 1 : 0번째 행 0번째 열 출력, 그냥 출력


## 2차원 배열에서의 Slicing

- a[0:2] -> [[1, 2, 3, 4], [5, 6, 7, 8]] : 2차원 배열 전체 출력
- a[0:2][0] -> [1, 2, 3, 4] : 1차원 형태로 출력
- a[0:2][0][1:] -> [2, 3, 4] : [1, 2, 3, 4] 중에서 1번째부터 끝까지 Slicing

```python
print(a[0:2][0][1:]) # [2, 3, 4]
```

```python
a
```

```python
print(a[::2])  # 0번째 행부터 2 간격으로 출력, 이때 출력은 2차원
print(a[::-1]) # 행에 대해 역순으로 출력
print(a[1::-1]) # [[5, 6, 7, 8], [1, 2, 3, 4]]
print()
print(a[::2][1]) # [[1, 2, 3, 4], [9, 10, 11, 12]]에 대해 첫 행 -> (a[::2])[1] 개념으로 봐야함 -> 차원 하나 줄어듦
```

```python
print(a[::2][::2]) # [[1, 2, 3, 4]] -> 이때는 2차원으로 나옴 (Slicing하면 차원에는 영향을 안 미침)
```
