```python
import pandas as pd
import numpy as np
s = pd.Series([1, 2, 3], index = list('abc'))
print(s)
print(s.values)
print(s.index)
```

```python
a = pd.Series({'a' : 'abc', 'b' : 1})
a
#사전형으로 하면 key가 index, value는 그대로 value
```

```python
a = pd.Series(['a', 'b', 'c'], index = range(1, 4))
print(a[1])
print(a[1:2]) #slicing : end-1까지 출력
print(a[[1, 2]]) #fancy indexing : 입력한 리스트 key값 그대로
```

```python
b = pd.Series([3, 4, 5], index = list('abc'))
print(b)
print(b['a']) #한 값을 지정하면 그 value만 출력
print(b['a':'b']) #Slicing 형태면 index값까지 같이 출력
print(b[['a', 'c']]) 
```

```python
c = pd.Series([5, 6, 7], index = [1, 2, 1])
c[1]
```

```python
A = pd.Series(['ab', 'ba', 'ab', 'ba', 'ab'], index = [1, 2, 2, 1, 3])
print(A[2])
print(A[A.values == 'ab'])
print('*' * 30)
```

```python
A = pd.Series(list('abc') * 2, 
              index = [k for j in [[i] * i 
                                            for i in range(1, 4)] 
                       for k in j])
print(A[-1:-6:-1])
print(A[0:5])

'''
k for j in [[i] * i for i in range(1, 4) 
                            for k in j]


'''
```
