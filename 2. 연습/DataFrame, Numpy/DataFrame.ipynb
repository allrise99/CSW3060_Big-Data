```python
a = [1, 2, 3]
b = [4, 5.0, 6]
c = [True, False, True]
d = ['A', 'B', 'C']
DF = pd.DataFrame([a, b, c, d])
DF
```

```python
'''
a = [1, 2, 3]
b = [4, 5.0, 6]
c = [True, False, True]
d = ['A', 'B', 'C']
'''
DF = pd.DataFrame({'A' : a, 'B' : b, 'C' : c, 'D' : d}, index = list('abc'))
DF
```

```python
#.loc : index, column name 기준으로 값을 뽑아냄
DF.loc['b', 'A']
```

```python
#.iloc : 행, 열 번호 기준으로 값을 뽑아냄
DF.iloc[2, 3]
```

```python
# .loc[ : , : ] - dataframe 출력
DF.loc[['c', 'a'], ['D', 'D', 'A']]
```

```python
# .iloc [ : , : ] - dataframe 출력
DF.iloc[1:2, 1:2]
```
