```python
import numpy as np

b = np.arange(1, 9).reshape(2, 4)
print(b)
```

```python
b[0:2] = np.arange(102, 110).reshape(2, 4)
print(b)
```

```python
b[0:2, 0:2] = np.arange(15, 19).reshape(2, 2)
b
```

```python
b[0:2, 0:2] = 16
b
```
