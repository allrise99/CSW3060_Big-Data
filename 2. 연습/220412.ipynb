```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [5, 2, 4])
```

```python
plt.plot([5, 2, 4]) #x축에 임의로 [0, 1, 2] 할당됨
```

```python
import numpy as np
plt.plot(np.arange(0, 5, 2) * np.arange(3)) 
#[0, 1, 2] : x축
#[0, 2, 4] * [0, 1, 2] = [0, 2, 8] (y축)
```

```python
plt.plot(np.arange(6).reshape(2, 3))

#[[0, 1, 2], [3, 4, 5]]
# line 1 : [0, 1] : x [0, 3] : y / line 2 : [1, 4] / line 3 : [2, 5]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
a = pd.Series(np.arange(0, 5, 2) * np.arange(3), index = range(1, 4))
a.index.name = 'x'
a.plot()

#index = [1, 2, 3]  : x축
#[0, 2, 4] * [0, 1, 2] = [0, 2, 8] : y축
```

```python
plt.plot(range(6))
pd.Series(np.arange(6,0,-1), index = range(1, 7)).plot()
```

```python
a = pd.DataFrame({'A' : dict([(1, 3), (3, 4), (5, 1)]), 
                  'B' : dict([(1, 1), (3, 2), (5, 5)])})

a.index.name = "X"
a.columns.name = "Class"
a.plot() #.plot() 메소드를 사용해야 해당 내용이 다 들어감
```

```python
plt.plot(range(5), range(6, 1, -1), '.b')

#[0, 1, 2, 3, 4] : x / [6, 5, 4, 3, 2] : y
```

```python
plt.plot(range(5), range(6, 1, -1), color = 'r', 
         linestyle = '-.', marker = 'o')
```

```python
plt.plot(range(5), range(6, 1, -1), 'k-',
         range(5), range(6, 1, -1), 'ro',
         range(5), range(7, 2, -1), 'g--',
         range(5), range(7, 2, -1), 'bo'
)

# plt.plot에서는 하나의 그림으로 합쳐서 표시가 가능
# 하지만 .plot() 메소드에서는 별개의 그림으로 나옴
```

```python
plt.plot(np.arange(0, 4, 0.5), list(range(4)) * 2, 'ko-',
         np.arange(0, 4, 0.5), list(range(4, 0, -1)) * 2, 'ko--'
)
```

```python
a = pd.DataFrame({'A' : list(range(4)) * 2, 
                  'B' : list(range(4, 0, -1)) * 2}, 
                 index = np.arange(0, 4, 0.5))

'''
     A  B
0.0  0  4
0.5  1  3
1.0  2  2
1.5  3  1
2.0  0  4
2.5  1  3
3.0  2  2
3.5  3  1
'''
# DataFrame에서 value 값은 y값
# index 값은 x값
a.index.name = 'size'
a.columns.name = 'time'
a.plot(linestyle = '--', marker =  'o')
```
