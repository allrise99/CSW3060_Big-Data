```python
import pandas as pd
a = pd.read_csv('Fame.txt', sep = '\t', index_col = 0) 
print(a)
```

```python
print(a.index)
```

```python
import math

for i in range(0, math.ceil(len(a.Name)/12)) :
    print((i*12), (i+1)*12)
```

```python
import math, time
from GoogleNgrams import Ngrams

b = []
for i in range(0, math.ceil(len(a.Name)/12)) :
    b.append(Ngrams(', '.join(a.Name[(i*12):((i+1)*12)])))
    time.sleep(1) #Ngramviewer가 한 IP에서 여러 번의 요청을 차단함 

c = pd.concat(b, axis = 1)
```

```python
d = []
for i in range(0, c.shape[-1]) :
    j = a['Birth.Year'][i] - 1800 
    Series = c.iloc[j:(j+100), i] #출생년도부터 100년만의 data 출력
    Series.index = range(0, 100) #index = 0 ~ 99
    d.append(Series)
e = pd.concat(d, axis = 1)
```

```python
e.to_csv('cohort.txt', sep = '\t')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
e.plot(legend = False)
```

```python
#plt.plot -> legend는 출력 안 됨

plt.plot(e, color = 'gray', linewidth = 1, alpha = 0.3, label = '_nolegend_')
plt.plot(e.median(axis=1), linewidth = 3, color = 'orange', label = 'median') #중앙값
plt.plot(e.mean(axis=1), linewidth = 3, color = 'skyblue', label = 'mean') # 산술평균
plt.legend()

'''
멀리서 보면 공통적인 특성을 발견할 수 있다.
비정규분포 -> 중앙값을 사용하는 것이 적절
'''
```
