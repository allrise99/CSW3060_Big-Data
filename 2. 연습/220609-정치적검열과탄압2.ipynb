```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('EU_painters_DF.txt', sep = "\t", encoding = "cp949", 
                header = [0, 1], index_col = 0)
a
```

```python
a.groupby(level = 'Corpus', axis = 1).median().plot()
plt.axvspan(1933, 1945, facecolor = 'grey', alpha = 0.3)
plt.annotate('', xy = (1945, 1.2e-7), xytext = (1933, 1.2e-7), 
             horizontalalignment = 'center',
             arrowprops = {'arrowstyle' : '<->', 'linewidth' : 2})
plt.annotate('나치 정권', ((1933+1945)/2, 1.25e-7), 
             horizontalalignment = 'center',
             size = 12)
plt.title('Median')
```

    - German은 감소폭이 더 큼. 왜 그럴까?
    - 나치의 화가들에 대한 검열 : 나치 정권 전에도 이들의 명성이 감소하는 추세
        - 왜 그런가?

- 상관성 vs 인과성 : 데이터는 상관적 지식만 알려줌.

```python
import numpy as np

plt.rcParams['figure.figsize'] = (5, 6)
plt.subplot(3, 1, 1)
plt.plot(a['German'])
plt.axvspan(1933, 1945, facecolor = 'grey', alpha = 0.3)
plt.title('German Corpus')
plt.ylim([0, np.array(a).max()+1e-8])

plt.subplot(3, 1, 2)
plt.plot(a['French'])
plt.axvspan(1933, 1945, facecolor = 'grey', alpha = 0.3)
plt.title('French Corpus')
plt.ylim([0, np.array(a).max()+1e-8])

plt.subplot(3, 1, 3)
plt.plot(a['American English'])
plt.axvspan(1933, 1945, facecolor = 'grey', alpha = 0.3)
plt.title('American English Corpus')
plt.ylim([0, np.array(a).max()+1e-8])
plt.tight_layout()
```
